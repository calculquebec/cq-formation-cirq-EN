# Quantum Fourier transform (QFT)

Given a periodic function, find it's period.

Example: cosine function



Example: more complex function. Period is not always evident.



Many data points are required to find the period of complex functions. Classically, this is a difficult problem that increases exponentially with the number of data points required to describe the period. With a quantum computer, the period can be found much more rapidly.

QFT finds the period of a function by a change of basis from the computational basis to the Fourier (Hadamard) basis.



where $N = 2^n$, $n$ is the number of qubits, and $\omega_N^nk$ = $e^\frac{2\pi i}{N}$ is the rotation (phase).

QFT for one qubit is the Hadamard gate.

## QFT circuit

This circuit implements QFT in the reverse order of the qubits and therefore a swap is required to output the correct ordering.

Why use a controlled unitary rotation?
- the phase is qubit dependent
- all possible combinations of 0s and 1s must be considered to apply the unitary rotation to the proper qubits
