# Factoring and Shor's algorithm

## Description of Shor's algorithm:
- Search for the prime factors of an integer N, which is a product of two prime numbers: $N = pq$, where $p$ and $q$ are prime numbers.
- RSA encryption uses a public key that is a product of 2 large prime numbers.
- Factoring N is a difficult problem on a classical computer.

## Protocol for Shor's algorithm: $N = pq$

1. Pick a number $a$ that is coprime witn $N$. Coprime means that $a$ has no common factors with $N$. For this example, we will factor 15 with 13 as a coprime.
2. Find the period $r$ of the function $f_{a,N}(x)$ = $a^r(mod N)$. We search for the smallest $r$ such that $a^r=1(mod N)$, *ie.* we look for the first time that the period goes back to 1.



3. If $r$ is even, then we can proceed, else we find another $a$. In this case, $r$ = 4. It is even, so we can proceed.

4. Quantum circuit to factor $N = 15 = [1 1 1 1]$, *ie.* 4 qubits



We have created a unitary oracle (recall the *Introduction to Quantum Computing* workshop).



5. We measure the bottom register of qubits to obtain 1, 13, 4, 7.
6. We apply an inverse QFT and then measure the input register to obtain the period.
7. Using a classical computer, we compute the two prime factors.
